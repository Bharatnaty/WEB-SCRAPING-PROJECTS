```python
import bs4
import pandas as pd
import csv
import requests
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url="https://www.worldometers.info/coronavirus/countries-where-coronavirus-has-spread/"
soup=requests.get(url)
soup=bs4.BeautifulSoup(soup.text,"html.parser")
soup=soup.find("div",class_="table-responsive")
soup=soup.find("table",class_="table table-striped table-bordered")
soup=soup.find_all("tr")[1:]
data=[]
for tr in soup:
  n=tr.text.strip().split("\n")
  data.append(n)
columns=["Country","Cases","Deaths","Region"]
data=pd.DataFrame(data,columns=columns)
```

```python
data=data.head(6)
data=data.sort_values(by=["Deaths"],ascending=True)
data
```

```python

plt.scatter(data["Country"].head(6),data["Cases"].head(6))
plt.title("No of cases in world")
plt.xlabel("Country")
plt.ylabel("Cases")
plt.show()
```

```python
plt.barh(data["Country"].head(6),data["Cases"].head(6))
plt.title("No of cases by country")
plt.xlabel("Country")
plt.ylabel("Cases")
plt.show()
```

```python
plt.plot(data["Country"].head(6),data["Deaths"].head(6))
plt.xlabel("country")
plt.ylabel("deaths")
plt.title("deaths count")
plt.show()
```

```python
plt.hist(data["Region"])
plt.xlabel("Country")
plt.ylabel("Region Affectes")
plt.title("region/Continent affected by Virus")
plt.show()
```

```python
sns.lineplot(data["Country"],data["Deaths"])
plt.show()
```
